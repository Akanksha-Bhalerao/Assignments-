μ=$1,000+$5(600)=$4,000

```python
import math
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
#Given Data
x_bar = 3050           # Sample mean (Rs.)
X = 600                # Units produced
mu = 1000 + 5 * X      # Theoretical mean from model
sigma = 5 * 25         # Standard deviation
n = 25                 # Sample size
alpha = 0.05           # Significance level
```

```python
# mean
mu = 1000 + 5 * X
mu
```

```python
# Standard deviation
sigma = 5 * 25
sigma
```

2. Calculate the Test Statistic

The population standard deviation (σ) is derived from the model as 5×25.


$$
Z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}} = \frac{3,050 - 4,000}{\frac{5 \times 25}{\sqrt{25}}} = \frac{-950}{25} = -38.00
$$



```python
#Calculate Test Statistic (Z)
z = (x_bar - mu) / (sigma / math.sqrt(n))
z
```

3. Determine the Critical Value

Using a significance level (α) of 0.05 for a right-tailed Z-test, the critical value is the Z-score that leaves 0.95 area to its left.

Z
crit
​
 =1.645

```python
#Critical Value for one-tailed test
z_critical = norm.ppf(1 - alpha)
z_critical
```

```python
#Decision
if z > z_critical:
    decision = "Reject Null Hypothesis (H₀)"
    result = "There is enough evidence that weekly costs are higher."
else:
    decision = "Fail to Reject Null Hypothesis (H₀)"
    result = "No evidence that costs are higher than the model predicts."

print(f"Decision: {decision}")
print(f"Conclusion: {result}")
```

```python
# 95% Confidence Interval for Mean
confidence = 0.95
ci_lower, ci_upper = stats.norm.interval(0.95, loc=x_bar, scale=sigma/math.sqrt(n))
print(f"\n{int(0.95*100)}% Confidence Interval for Mean Weekly Cost: ({ci_lower:.2f}, {ci_upper:.2f})")

if mu > ci_upper:
    print("→ Theoretical mean (4000) lies ABOVE this interval, meaning actual costs are lower.")
elif mu < ci_lower:
    print("→ Theoretical mean (4000) lies BELOW this interval, meaning actual costs are higher.")
else:
    print("→ Theoretical mean (4000) lies WITHIN this interval, meaning model is consistent.")
```
