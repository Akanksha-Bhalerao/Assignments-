# Exercise 1: Prime Numbers

```python
def is_prime(number):
    # A prime number must be greater than 1
    if number <= 1:
        return False
    # Check for factors from 2 up to the square root of the number
    # We only need to check up to int(number**0.5) because any factor
    # larger than the square root must have a corresponding smaller factor.
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

num_check_1 = 17
num_check_2 = 15
print(f"Is {num_check_1} prime? {is_prime(num_check_1)}")
print(f"Is {num_check_2} prime? {is_prime(num_check_2)}")
# Expected Output: True, False
```

# Exercise 2: Product of Random Numbers

```python
import random

def quiz_product():
    # Generate two random integers between 1 and 10
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    correct_product = num1 * num2

    # Ask the user for their answer
    # CORRECTED LINE: The prompt should NOT contain the answer.
    print(f"\nWhat is the product of {num1} and {num2}?")

    try:
        # CORRECTED LINE: The prompt should only ask for the input.
        user_answer = int(input("Your answer: "))

        # Check the answer
        if user_answer == correct_product:
            print(" Correct! Well done.")
        else:
            print(f" Incorrect. The correct answer is {correct_product}.")

    except ValueError:
        print("Invalid input. Please enter a whole number.")

# --- Demonstration (Requires user interaction) ---
# REMOVE THE # TO RUN THE CODE
quiz_product()
```

# Exercise 4: Word counter

```python
def word_counter(input_text):
    word_counts = {}

    # 1. Clean the text: Convert to lowercase to count 'This' and 'this' together.
    # We also remove common punctuation marks for a better word count.
    cleaned_text = input_text.lower().replace('.', '').replace(',', '').strip()

    # 2. Split the text into a list of words
    words = cleaned_text.split()

    # 3. Count the frequency of each word
    for word in words:
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1

    return word_counts

# --- Demonstration ---
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
word_frequencies = word_counter(input_text)

print("\n--- Word Frequencies ---")
for word, count in word_frequencies.items():
    print(f"'{word}': {count}")

# Expected Output (matching example): 'this': 2, 'is': 1, 'a': 1, 'sample': 1, 'text': 2, 'will': 1, 'be': 1, 'used': 1, 'to': 1, 'demonstrate': 1, 'the': 1, 'word': 1, 'counter': 1
```

# Exercise 5: Check for Palindrome

```python
import re # Used for removing non-alphanumeric characters

def is_palindrome(input_string):
    """Checks if a string is a palindrome, ignoring spaces, punctuation, and capitalization."""

    # 1. Normalize the string: Convert to lowercase
    normalized_string = input_string.lower()

    # 2. Remove non-alphanumeric characters (spaces, punctuation, etc.)
    # The sub function substitutes all matches of the pattern (anything not a letter or number) with an empty string.
    cleaned_string = re.sub(r'[^a-z0-9]', '', normalized_string)

    # 3. Check for palindrome: Compare the cleaned string with its reverse
    # [::-1] is Python's slicing notation for reversing a string.
    return cleaned_string == cleaned_string[::-1]

# --- Demonstration ---
print("\n--- Palindrome Check ---")
print(f"Input: 'racecar', Output: {is_palindrome('racecar')}")
print(f"Input: 'A man, a plan, a canal: Panama', Output: {is_palindrome('A man, a plan, a canal: Panama')}")
print(f"Input: 'hello world', Output: {is_palindrome('hello world')}")
# Expected Output: True, True, False
```
