```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Basic Data Exploration
The "Wine" dataset contains 178 entries and 14 features. All columns are non-null and have appropriate numerical data types.



```python
# Load the dataset (Task 1)
df = pd.read_csv('/content/wine.csv')
df
```

```python
# Separate features (X) from the target (y - 'Type' is used for evaluation/context)
X = df.drop('Type', axis=1)
y = df['Type']
```

```python
# 1. Identify features to plot (excluding the target 'Type')
features = df.columns.drop('Type')

# 2. Set up the visualization grid
plt.figure(figsize=(16, 12))
plt.suptitle('Distribution of Individual Features in Wine Dataset', fontsize=16, y=1.02)

# 3. Plot a histogram for each feature
for i, col in enumerate(features):
    plt.subplot(4, 4, i + 1) # 4 rows, 4 columns
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'{col}', fontsize=12)
    plt.xlabel('')
    plt.ylabel('Frequency', fontsize=10)

plt.tight_layout()
plt.show() # Display the chart
```

```python
# Task 2: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
# EDA: Feature Correlation Heatmap (Task 1)
plt.figure(figsize=(10, 8))
sns.heatmap(X_scaled_df.corr(), annot=False, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap (Task 1)')
plt.show()
```

```python
# Implement PCA and determine optimal components (Task 2)
pca_all = PCA()
pca_all.fit(X_scaled)
```

```python
# Scree Plot / Cumulative Explained Variance Plot (Task 2)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1, len(pca_all.explained_variance_ratio_) + 1), pca_all.explained_variance_ratio_, marker='o', linestyle='--')
plt.title('Scree Plot (Task 2)')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.subplot(1, 2, 2)
plt.plot(range(1, len(pca_all.explained_variance_ratio_) + 1), pca_all.explained_variance_ratio_.cumsum(), marker='o', linestyle='-')
plt.title('Cumulative Explained Variance (Task 2)')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Variance Explained')
plt.tight_layout()
plt.show()
```

```python
# Transform the original dataset into 2 principal components (Task 2)
# Choosing 2 components for visualization, which captures over 55% of the variance
n_components = 2
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)
X_pca_df = pd.DataFrame(data=X_pca, columns=[f'PC{i+1}' for i in range(n_components)])
```

```python
# Store explained variance
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()[n_components-1]
```

```python
# --- K-Means Clustering (K=3) ---
K = 3
```

```python
# Task 3: Clustering with Original Data
kmeans_original = KMeans(n_clusters=K, random_state=42, n_init='auto')
clusters_original = kmeans_original.fit_predict(X_scaled)
```

```python
# Evaluate Clustering (Task 3)
score_silhouette_original = silhouette_score(X_scaled, clusters_original)
score_davies_bouldin_original = davies_bouldin_score(X_scaled, clusters_original)
```

```python
# Task 4: Clustering with PCA Data
kmeans_pca = KMeans(n_clusters=K, random_state=42, n_init='auto')
clusters_pca = kmeans_pca.fit_predict(X_pca)
```

```python
# Evaluate Clustering (Task 4)
score_silhouette_pca = silhouette_score(X_pca, clusters_pca)
score_davies_bouldin_pca = davies_bouldin_score(X_pca, clusters_pca)
```

```python
# Visualization of PCA-Clustering Results (Task 4)
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=X_pca_df['PC1'],
    y=X_pca_df['PC2'],
    hue=clusters_pca,
    palette='viridis',
    legend='full',
    alpha=0.8
)
plt.title('K-Means Clustering on PCA-Transformed Data (Task 4)')
plt.xlabel(f'Principal Component 1 ({explained_variance_ratio[0]*100:.1f}% Variance)')
plt.ylabel(f'Principal Component 2 ({explained_variance_ratio[1]*100:.1f}% Variance)')
plt.show()
```

```python
# Final results for comparison (Task 5 data)
results = {
    'Explained_Variance_2_PC': cumulative_explained_variance,
    'Silhouette_Original': score_silhouette_original,
    'Davies_Bouldin_Original': score_davies_bouldin_original,
    'Silhouette_PCA': score_silhouette_pca,
    'Davies_Bouldin_PCA': score_davies_bouldin_pca
}
results
```

* Feature Distributions Visualization

The histograms show that many features, such as Magnesium and Color, are right-skewed, with a few features like Malic appearing more uniformly distributed. This difference in scale and distribution confirms the necessity of Standardization before applying PCA or distance-based clustering.

* Clustering Visualization (Original Data)

K-Means clustering with $K=3$ (matching the actual number of wine classes) was applied to the Standard Scaled original data. The visualization below uses two features, Proline and Flavanoids, which typically show good separation.The resulting clusters show some degree of overlap, particularly between the blue and green clusters, indicating that the K-Means algorithm on the full, unreduced feature space struggles to find perfectly distinct boundaries. This emphasizes the potential value of PCA in preprocessing.

Analysis:
 Clustering quality significantly improved after applying PCA. Both the Silhouette Score and Davies-Bouldin Index show better cluster separation and compactness for the PCA-transformed data. This indicates that the 2 principal components, which capture 55.4% of the total variance, retained the most critical information necessary to distinguish the wine types while filtering out noise and redundancy from the highly correlated original features. The trade-off is a loss of interpretability in exchange for superior clustering performance and computational efficiency.

Conclusion and Insights

Conclusion:
PCA is an effective preprocessing step for K-means clustering on the Wine dataset. It successfully decorrelates features, leading to better-defined clusters (Silhouette Score â†‘, Davies-Bouldin Index â†“).

Practical Implications:
PCA is recommended when the priority is model quality and efficiency in datasets with high dimensionality and feature correlation. It is the preferred method for exploratory data analysis and visualization. If direct feature interpretability is required, the original features must be used, accepting the potential degradation in clustering performance.

Recommendations:

Use PCA when: The primary goal is model performance (like clustering quality) or visualization, and the dataset has a large number of potentially correlated features. It is the recommended approach for exploratory analysis where feature relationships are complex.

Use Original Data when: Direct feature interpretability is paramount. If the business objective requires explaining exactly which feature values (e.g., specific alcohol or phenol levels) define a cluster, using PCA is a hindrance, and one must tolerate the lower clustering quality of the original space.
