```python
# Import essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.dpi'] = 100
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/Cardiotocographic.csv")
df
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
median_values = df.median()
df.fillna(median_values, inplace=True)
median_values
```

```python
numerical_cols_to_correct = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width', 'NSP']
numerical_cols_to_correct
```

```python
df.shape
```

```python
df[numerical_cols_to_correct] = df[numerical_cols_to_correct].clip(lower=0)
```

```python
summary_stats = df.describe().T[['mean', '50%', 'std', '25%', '75%']]
```

```python
summary_stats = summary_stats.rename(columns={'50%': 'median',
                                              '25%': 'Q1 (25th perc.)', # ðŸ‘ˆ This is the correct column name
                                              '75%': 'Q3 (75th perc.)'})
```

```python
summary_stats['IQR'] = summary_stats['Q3 (75th perc.)'] - summary_stats['Q1 (25th perc.)']
```

```python
summary_stats
```

```python
# Assuming df is your main DataFrame
key_features = ['LB', 'ASTV', 'MLTV']

# 1. Reshape the data from wide format to long format (melting)
# This stacks the 'key_features' into a single 'Variable' column
df_melted = pd.melt(df[key_features], var_name='Variable', value_name='Value')

# 2. Use Seaborn's FacetGrid to plot all distributions simultaneously
g = sns.FacetGrid(df_melted, col='Variable', sharex=False, sharey=False, height=4)
g.map(sns.histplot, 'Value', kde=True, bins=30, color='skyblue')

# Add titles and adjust layout
g.set_axis_labels("Value", "Frequency")

plt.tight_layout()
plt.show()

```

```python
# --- Boxplot
plt.figure(figsize=(15, 6))
sns.boxplot(data=df[key_features], orient="h", palette="Pastel1")
plt.title('Boxplots of Key CTG Features')
plt.show()
```

```python
# **Correlation Heatmap**
plt.figure(figsize=(12, 10))
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a mask for the upper triangle
mask = np.triu(correlation_matrix)

# Plot the heatmap
sns.heatmap(correlation_matrix,
            annot=False, # Set to True to display correlation values
            cmap='coolwarm',
            fmt=".2f",
            linewidths=.5,
            mask=mask,
            cbar_kws={'shrink': .8})
plt.title('Correlation Heatmap of CTG Variables')
plt.show()
```

```python
# **Scatter Plot: Relationship between Baseline FHR (LB) and Decelerations (DL)**
plt.figure(figsize=(8, 6))
sns.scatterplot(x='LB', y='DL', data=df, alpha=0.6)
plt.title('Baseline FHR vs. Late Decelerations (DL)')
plt.xlabel('Baseline Fetal Heart Rate (LB)')
plt.ylabel('Late Decelerations (DL)')
plt.show()
```

#  Conclusion

## Summary of Key Insights:

1.  **Data Quality:** The initial dataset required significant cleaning, including converting potential string/object types to numeric and handling a notable number of negative or inconsistent data points by setting them to zero.
2.  **FHR Baseline (LB):** The FHR baseline is normally distributed around a mean of [133.340204] bpm, indicating most fetuses in the sample are within a healthy range.
3.  **Variability:** The statistical summary showed a large spread (high IQR) in both Short-Term (ASTV) and Long-Term Variability (ALTV). This suggests significant differences in fetal heart rate health across the monitored individuals.
4.  **Strongest Relationships:** The correlation heatmap revealed a strong positive relationship between [ALTV and MLTV], suggesting these measures track similar aspects of FHR stability. Conversely, there is often a strong negative relationship between baseline FHR and some variability measures, which is a known physiological effect.

## Recommendations for Further Analysis:

* **Classification Task:** The ultimate goal of CTG data is often to classify fetal status (e.g., normal, suspicious, pathological). The cleaned data should now be used to train a machine learning model to predict the **NSP** (Fetal State Class) variable.
* **Feature Engineering:** Since the data is highly correlated, consider using **Principal Component Analysis (PCA)** to reduce dimensionality and combine related features before modeling.
