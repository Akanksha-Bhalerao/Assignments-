# Standardization (Z-score)

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Define the numerical columns
numerical_cols = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

# Line 1: Initialize the StandardScaler
scaler = StandardScaler()

# Line 2: Fit the scaler to the data and transform the columns in place
df_numerical_cols = scaler.fit_transform(df[numerical_cols])
```

```python
df_numerical_cols
```

The goal of Z-score normalization is to scale all numerical data to a common range without distorting differences in the ranges of values.

Process: Z-score standardization transforms the data so that it has a mean of 0 and a standard deviation of 1.

Why it's needed: Standardization is crucial for distance-based machine learning algorithms (like K-Nearest Neighbors or k-means clustering). It prevents features with naturally larger values (e.g., Net Sales Value) from disproportionately influencing the model's results compared to features with smaller values (e.g., Discount Rate (%)).

# One-Hot Encoding

```python
# Identify categorical columns to encode
cols_to_encode = ['Day', 'SKU', 'City', 'BU', 'Brand']

# Apply one-hot encoding
# drop_first=True prevents multicollinearity by dropping one binary column for each group
df_encoded = pd.get_dummies(df, columns=cols_to_encode, drop_first=True)
```

```python
df_encoded
```

```python
df_encoded.describe()
```

One-Hot Encoding is a vital data preprocessing step for two main reasons:

Model Requirement: Machine learning algorithms (such as linear regression, neural networks, or support vector machines) can only process numerical data. Categorical text values (like 'RealU' or 'FMCG') cannot be directly used in mathematical models.

Avoid Misinterpretation: Converting categories to simple numerical labels (e.g., Brand A=1, Brand B=2) implies an ordered relationship, which is false for nominal data. One-Hot Encoding solves this by creating a new binary (0 or 1) column for each category, ensuring the model sees categories as distinct, unordered entities.
